```python
import numpy as np
import pandas as pd

import geopandas as gpd
```

Read data

```python
SHAPE_FILE = "data/gadm36_PHL_shp/gadm36_PHL_3.shp"
shapes = gpd.read_file(SHAPE_FILE)
```

```python
shapes.head()
```

```python
sample = pd.DataFrame(data = zip([1,2,3], [10.541173972241738, 14.54136684853697, 8.041173972241738], 
                     [122.83989386089448, 121.03749818490371, 123.0374981840371]), 
             columns=['id', 'latitude', 'longitude'])


def simple_sjoin(df, shapes):
    #convert to GeoDataFrame and add CRS 4326
    df = gpd.GeoDataFrame(
        df, geometry=gpd.points_from_xy(df['longitude'], df['latitude']))
    df.set_crs(epsg=4326, inplace=True)
    
    df_joined = gpd.sjoin(shapes, df)
    return df_joined

def sjoin_point(lat, lon, shapes):
    df = pd.DataFrame(data = [[lat,lon]], columns=['latitude', 'longitude'])
    
    df_joined = simple_sjoin(df, shapes)
    
    return df_joined
```

```python
sjoin_point(1,1,shapes)
```
